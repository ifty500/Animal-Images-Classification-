```python
import numpy as np
import keras
from keras.layers import Dense
from keras.models import Sequential
from tensorflow import random
from tensorflow.keras.applications.vgg16 import preprocess_input
import matplotlib.pyplot as plt
from glob import glob
from keras.callbacks import EarlyStopping
```

```python
vgg_model = keras.applications.vgg16.VGG16()
```

```python
vgg_model.summary()
```

```python
last_layer = str(vgg_model.layers[-1])
np.random.seed(42)
random.set_seed(42)

model= Sequential()

for layer in vgg_model.layers:
    if str(layer)!= last_layer:
        model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable= False
```

```python
model.summary()
```

```python
folders = glob('/kaggle/input/animal-images/animal_dataset_intermediate/train/*')
```

```python
model.add(Dense(len(folders), activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

generate_train_data = ImageDataGenerator(rescale = 1./255,
                                         validation_split = 0.2,
                                         height_shift_range = 0.2,
                                         width_shift_range = 0.2,
                                         shear_range = 0.2,
                                         zoom_range = 0.2,
                                         horizontal_flip = True)
val_datagen = ImageDataGenerator(rescale = 1.0/255,validation_split = 0.2)

generate_test_data = ImageDataGenerator(rescale =1./255)
```

```python
training_dataset = generate_train_data.flow_from_directory('/kaggle/input/animal-images/animal_dataset_intermediate/train/',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 subset="training",
                                                 class_mode = 'categorical')

# test_datasetset = generate_test_data.flow_from_directory('/kaggle/input/animal-images/animal_dataset_intermediate/test/',
#                                             target_size = (224, 224),
#                                             batch_size = 32,
                                            
#                                             class_mode = 'categorical')

val_generator = val_datagen.flow_from_directory("/kaggle/input/animal-images/animal_dataset_intermediate/train/",
                                                subset = "validation",
                                                    batch_size = 32,
                                                    class_mode = 'categorical', 
                                                    target_size = (224, 224))
```

```python
es_callback = EarlyStopping(monitor='val_loss', patience=2,
                                               restore_best_weights=True,
                                               verbose=1)
history=model.fit(training_dataset,
                         steps_per_epoch = len(training_dataset),
                         epochs = 20,
                         validation_data = val_generator,
                         validation_steps = len(val_generator),
                         callbacks = [es_callback],
                         shuffle=False)
```

```python
training_dataset.class_indices
```

```python
# plot the loss
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
